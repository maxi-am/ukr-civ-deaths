### Intro

This notebook is devoted to vizualisation the number of victims through civilians in conflict in Ukraine during 2014-2022. It was prepared for my Russian friends - social activists.

The data for this vizualisation taken from the official website of the Office of the United Nations Human Rights Office of High Commissioner.

Link for 2014-2021 years data (page 16):
https://www.ohchr.org/sites/default/files/Documents/Countries/UA/32ndReportUkraine-ru.pdf 

Link for 2022 years data:
https://www.ohchr.org/en/press-releases/2022/07/ukraine-civilian-casualty-update-25-july-2022

Reports for 2022 years are updating with new numbers several times at month. 

```python
import pandas as pd
from datetime import date
import matplotlib.pyplot as plt
```

<b>The cell below is for manual entering the data about numbers of victims.</b>

```python
start_day=date(2022,2,24)
report_day=date(2022,7,24)

days=(report_day-start_day).days

last_deaths_info=5237

years=['2014','2015','2016','2017','2018','2019','2020','2021','2022']

deaths=[2084,954,112,117,55,27,26,15,last_deaths_info]
dict_of_lists={'years':years,'deaths':deaths}

deaths_data=pd.DataFrame(dict_of_lists)
```

```python
deaths_data.head()
```

```python
# calculating amount of deaths for 2014-2021 period

y8_deaths=deaths_data['deaths'].iloc[:-1].sum()
y8_deaths_str="Количество жертв 2014-2021: "+str(y8_deaths)

print(y8_deaths_str)
```

```python
# preparing visualization and annotations for better readability and representation

plt.style.use("bmh")
fig,ax = plt.subplots()

p1= ax.bar(deaths_data['years'], deaths_data['deaths'], color=['b','b','b','b','b','b','b','b','red'])
ax.set_xticks(deaths_data['years'])
ax.set_ylim(top=(deaths_data.iloc[8,-1]//500+2)*500)
ax.set_xticklabels(deaths_data['years'], rotation=90)
ax.set_ylabel("Кол-во смертей гражданских лиц")
ax.set_title("Данные мониторинговой группы ООН\nо жертвах конфликта в Украине за 2014-2022")

ax.bar_label(p1,fontsize=13,fontweight=700)
ax.annotate("Количество смертей среди",             
	xy=(3, 5),	xytext=(0.1, 2900),fontsize=13,fontweight=300) 
ax.annotate("гражданских за 2014-2021: ",             
	xy=(3, 5),	xytext=(0.1, 2600),fontsize=13,fontweight=300) 
ax.annotate(str(y8_deaths),             
	xy=(3, 5),	xytext=(4.8, 2600),fontsize=13,fontweight=700) 
ax.annotate(str('за ')+str(days)+str(' дней >>'),             
	xy=(3, 5),	xytext=(5.3, deaths_data.iloc[8,-1]-90),fontsize=11,fontweight=300)
ax.annotate(str('за 8 лет >>'),             
	xy=(3, 5),	xytext=(5.5, 3250),fontsize=11,fontweight=300,color='grey')
ax.annotate(str('_ _ _ _'),             
	xy=(3, 5),	xytext=(7.6, 3420),fontsize=11,fontweight=200)
#p1=ax.plot(years[:-1],deaths[:-1])
fig.set(figwidth=7,figheight=4.6)
plt.show()
```
